Importing the Dependencies.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import sklearn.svm as svm
from sklearn.metrics import accuracy_score
```

```python
diabetes_dataset = pd.read_csv('/content/diabetes_dataset.csv')
```

Splitting the dataset file into training and the testing dataset file.

```python
X = diabetes_dataset.drop(columns = 'Outcome', axis = 1)
Y = diabetes_dataset['Outcome']
```

```python
X.head()
```

```python
Y.head()
```

Data Standardization for making the better Predictions.

```python
scaler = StandardScaler()
```

```python
standardized_data_X = scaler.fit_transform(X)
```

```python
X = standardized_data_X
Y = diabetes_dataset['Outcome']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 3, stratify = Y)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

Training the model so that it can predict, whether the person is diabetic or non-diabetic.

```python
classifier = svm.SVC(kernel = 'linear').fit(X_train, Y_train)
```

```python
trained_model_prediction = classifier.predict(X_train)
trained_model_accuracy = accuracy_score(trained_model_prediction, Y_train)
```

```python
print(f'The accuracy score of the trained model is {trained_model_accuracy}')
```

```python
test_model_prediction = classifier.predict(X_test)
test_model_accuracy = accuracy_score(test_model_prediction, Y_test)
```

```python
print(f'The accuracy score of the test model is {test_model_accuracy}')
```

Making the Predictive Model.

```python
input_data =(0,137,40,35,168,43.1,2.288,33)
input_data_as_np_array = np.asarray(input_data)
input_data_reshaped = input_data_as_np_array.reshape(1, -1)
prediction = classifier.predict(input_data_reshaped)
if prediction[0] == 0:
  print('The person is Non-Diabetic')
else:
  print('The person is Diabetic')
```
